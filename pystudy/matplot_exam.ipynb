```python
# matplotlib = 데이터를 표현할 수 있게 해주는 파이썬 라이브러리
# figure => 큰 도화지
# subplots => 도화지 위에 나눠진 작은 칸들(한 도화지에 다양한 표, 그래프 그릴 수 있음)
```

```python
import matplotlib.pyplot as plt

# 기본스타일 적용
plt.style.use("default")

# y값만 전달 x축은 자동으로 0, 1, 2, 3...이 설정됨
plt.plot([2, 4, 6, 8])

plt.show()
```

```python
# x값도 지정 - 각 축의 기준값이 변경된 그래프 출력
x = [2, 4, 6, 8]
y = [1, 3, 5, 7]
plt.plot(x, y)
plt.show()
```

```python
x = range(11) # x : 1 ~ 10
y = range(5, 16) # y :5 ~ 15

# r : red, o : 동그라미 == 마커
plt.plot(x, y, "ro")
plt.show()
```

```python
fig = plt.figure() # 큰 도화지를 만듦 _크기 조정도 가능

# 2: 도화지를 2개의 행(row)으로 나눈다
# 1: 한 행에 1개의 열(column)을 가진다
# 1: 그 중에서 첫 번째 칸을 사용한다(위쪽 칸)
fg1 = fig.add_subplot(2, 1, 1) 
fg2 = fig.add_subplot(2, 1, 2) 

fg1.plot([1, 2, 3], [4, 5, 6])
fg2.plot([2, 4, 6], [1, 3, 7])

plt.show()
```

```python
# 예제
# 1. 도화지 생성
# 2. 도화지를 가로 방향으로 3개의 칸으로 나눈다 (1행 3열)
# 3. 그 중 첫 번째 칸과 세 번째 칸에만 그래프 영역을 만든다
# - 두 번째 칸은 비워둔다

pp = plt.figure()
pp1 = pp.add_subplot(1, 3, 1)
pp2 = pp.add_subplot(1, 3, 3)
plt.show()
```
